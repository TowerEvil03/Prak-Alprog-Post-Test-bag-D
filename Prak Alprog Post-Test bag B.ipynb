```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Kelas F_Student AI Usage.csv')

std_grade = df.groupby('education_level')['grades_after_ai'].std()

print(std_grade)

std_grade.plot(kind='bar')
plt.title('Standar Deviasi Nilai Setelah Menggunakan AI')
plt.xlabel('Jenjang Pendidikan')
plt.ylabel('Standar Deviasi')
plt.show()
```

```python
#Pada analisis ini dilakukan proses penyaringan (filtering) terhadap data responden
#yang memiliki jenjang pendidikan college dan memperoleh nilai akademik setelah
#menggunakan AI (grades_after_ai) di atas 80. Tujuan dari penyaringan ini adalah
#untuk mengidentifikasi kelompok mahasiswa yang menunjukkan performa akademik
#tinggi setelah memanfaatkan teknologi AI dalam proses pembelajaran.
#Hasil filtering menghasilkan sejumlah data mahasiswa yang memenuhi kriteria tersebut dan
#kemudian divisualisasikan menggunakan diagram batang (bar chart) untuk mempermudah analisis.

#Berdasarkan visualisasi yang diperoleh, dapat diamati distribusi mahasiswa
#berprestasi berdasarkan tujuan penggunaan AI. Hasil ini menunjukkan bahwa
#pemanfaatan AI sebagai alat bantu belajar dapat memberikan dampak positif
#terhadap pencapaian akademik mahasiswa. Dengan adanya AI, mahasiswa dapat
#memperoleh bantuan dalam memahami materi, mencari referensi, menyelesaikan
#tugas, maupun meningkatkan efektivitas belajar. Oleh karena itu, penggunaan AI
#yang tepat dan bertanggung jawab berpotensi membantu mahasiswa mencapai hasil
#akademik yang lebih baik.
```
