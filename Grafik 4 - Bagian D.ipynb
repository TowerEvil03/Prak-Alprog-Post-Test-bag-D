```python
import pandas as pd
import matplotlib as mt
import seaborn as sns
import matplotlib.pyplot as plt
import io

from google.colab import files
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['Kelas F_Student AI Usage (10).csv']))

plt.figure(figsize=(8, 6))
sns.boxplot(x='education_level', y='study_hours_per_day', data=df)
plt.title('Sebaran Jam Belajar Harian Berdasarkan Jenjang Pendidikan')
plt.xlabel('Jenjang Pendidikan')
plt.ylabel('Jam Belajar per Hari')
plt.show()
```
