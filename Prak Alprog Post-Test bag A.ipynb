```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])
```

```python
df.head()
```

```python
std_edu = (
    df.groupby('education_level')['grades_after_ai']
      .std()
      .reset_index()
)

print(std_edu)
```

```python
fig, ax = plt.subplots(figsize=(6,5))

ax.bar(
    std_edu['education_level'],
    std_edu['grades_after_ai']
)

for i, v in enumerate(std_edu['grades_after_ai']):
    ax.text(
        i,
        v,
        f'{v:.4f}',
        ha='center',
        va='bottom'
    )

plt.show()
```
