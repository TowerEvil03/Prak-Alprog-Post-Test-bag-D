```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv(list(uploaded.keys())[0])
```

```
df.head()
```

```
print(df.columns)
```

```
df['improvement'] = df['grades_after_ai'] - df['grades_before_ai']
```

```
df[['age','grades_before_ai','grades_after_ai','improvement']].head()
```

```
corr = df['age'].corr(df['improvement'])

print("Nilai Korelasi =", corr)
```

```
fig, ax = plt.subplots(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='age',
    y='improvement',
    ax=ax
)

ax.set_title('Korelasi Umur dengan Peningkatan Nilai Setelah AI')
ax.set_xlabel('Umur')
ax.set_ylabel('Peningkatan Nilai')

plt.show()
```
